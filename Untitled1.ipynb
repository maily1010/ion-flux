```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the files
data1 = np.loadtxt('sn4-TLCL2_1998.xvg')
data2 = np.loadtxt('sn4-TLCL2_1999.xvg')

# Extract the columns
x = data1[:, 0]
y = data2[:, 1]

# Create the scatter plot
plt.scatter(x, y, s=10, c='b', alpha=0.5)

# Set the axis labels and title
plt.xlabel('Column 1')
plt.ylabel('Column 2')

plt.title('Scatter Plot')

# Display the plot
plt.show()
```
